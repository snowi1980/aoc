## Puzzle 2-1

[Puzzle Day 2](https://adventofcode.com/2023/day/2)

```python
import string
import re

def read_file_to_string_array(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    # Strip newline characters from each line
    lines = [line.strip() for line in lines]
    return lines

# Example usage
file_path = './puzzle2-1.txt'
string_array = read_file_to_string_array(file_path)

max_cubes = {
  "red" : 12 ,
  "green" : 13,
  "blue" : 14,

}

# Function to convert a string to an integer
def convert_to_int(value):
    try:
        return int(value)
    except ValueError:
        return None  # Return None if the string cannot be converted to an integer

def get_game_id(s):
  match = re.search(r'\b\d+\b', s)
  if match:
      return int(match.group())
  return None

def extract_max_cubes(s,color):
    cubes = 0
    pattern = rf'(\d+)\s+{color}'
    matches = re.findall(pattern, s)
    if matches:
        for match in matches:
          if int(match) > cubes:
            cubes = int(match)
        return cubes
    return 0

def check_valid_game(s):
  id = get_game_id(s)
  game_data = f"{id}:"
  for cube in max_cubes:
    cubes = extract_max_cubes(s, cube)
    game_data += f"{cube} {cubes} ,"
    print(game_data)
    if cubes > max_cubes[cube]:
      return 0
  
  
  return id    

def check_game_power(s):
  id = get_game_id(s)
  game_data = f"{id}:"
  power = 1
  for cube in max_cubes:
    cubes = extract_max_cubes(s, cube)
    game_data += f"{cube} {cubes} ,"
    print(game_data)
    power *= cubes
  
  
  return power    


result = 0
power = 0
for s in string_array:
  
  valid_game = check_valid_game(s)
  if valid_game > 0:
    print(valid_game)
  result += valid_game

  power += check_game_power(s)
  

print(f"Result is {result}")
print(f"Power is {power}")
```
