### Common stuff

```python
import string

def read_file_to_string_array(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    # Strip newline characters from each line
    lines = [line.strip() for line in lines]
    return lines

def read_file_to_2d_char_array(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    # Convert each line to a list of characters and store in a 2D array
    char_array = [list(line.strip()) for line in lines]
    return char_array

class ProductNumber:
    def __init__(self, value, rownum, column, length):
        self.value = value
        self.rownum = rownum
        self.column = column
        self.length = length
```

### Puzzle 3-1

```python
file_path = './puzzle3-1.txt'
#string_array = read_file_to_string_array(file_path)
char_array = read_file_to_2d_char_array(file_path)

product_numbers = []

def create_product_numbers(row, rownum):
    value = ""
    length = 0
    column = 0

    for c in row:
        #print(f" Char:{c}")
        if(c.isdigit()):
          value += c
          length += 1
        else:
            if(length >= 1):
                product_numbers.append(ProductNumber(int(value), rownum, column-length, length))
                value = ""
                length = 0    
        column += 1
    
    if(length >= 1):
        product_numbers.append(ProductNumber(int(value), rownum, column-length, length))
    
        


    
def is_valid_product_number(product_number):
  #  print(f"Values: {product_number.value}, Row: {product_number.rownum}, Column: {product_number.column}, Length: {product_number.length}")

    startrow = max(0, product_number.rownum - 1)
    endrow = min(len(char_array), product_number.rownum + 2)

    startcol = max(0, product_number.column - 1)
    endcol = min(len(char_array[0]), product_number.column + product_number.length + 1)

  #  print(f"Checking row: {startrow} - {endrow}, column: {startcol} - {endcol}")
    #print("----------")
    text = ""
    valid = False
    for i in range(startrow, endrow):
        text += "\n"
        for j in range(startcol, endcol):
            if(not(char_array[i][j].isdigit() or char_array[i][j] == ".")):
              valid = True
            text += char_array[i][j]
        
    

    #print(f"{text}")

    if(valid):
        print(f"{product_number.value}")
        #print(f"Values: {product_number.value}, Row: {product_number.rownum}, Column: {product_number.column}, Length: {product_number.length}")
        return product_number.value
    else:
        return 0        

rownum = 0
for row in char_array:
    create_product_numbers(row, rownum)
    rownum += 1

max_value = 0
result = 0
for product_number in product_numbers:
    #print(f"{product_number.value}")
    #print(f"Values: {product_number.value}, Row: {product_number.rownum}, Column: {product_number.column}, Length: {product_number.length}")
#max_value += product_number.value
    result += is_valid_product_number(product_number)



print(f"Max Value: {max_value}")
print(f"Result: {result}")
assert result == 531932

#526341 wrong
#553825 correct
#589283 max value
```
